```python
from ultralytics import YOLO
```

```python
model=YOLO('yolov8n.pt')
```

```python
import cv2

cap=cv2.VideoCapture(0)

while cap.isOpened():
    OPEN,frame=cap.read()
    if not OPEN:
        print('camera is not open')
        break
        
    results=model(frame)
    for i in results:
        bbox=i.boxes.xyxy.numpy()
        confs=i.boxes.conf.numpy()
        class_ids=i.boxes.cls.numpy()
        
    for box,conf,cls in zip(bbox,confs,class_ids):
            x1,y1,x2,y2=map(int,box)
            bbox_color=(0,255,0)
            thickness=1
            cv2.rectangle(frame,(x1,y1),(x2,y2),(0,0,255),thickness)
            #class_names=model.names

            label=f'{model.names[cls]}:{conf:.2f}'
            cv2.putText(frame,label,(x1,y1-20),cv2.FONT_HERSHEY_SIMPLEX, 0.5,(0,0,0),2)
            
    cv2.imshow('video',frame)
    if cv2.waitKey(1) & 0xFF==ord('q'):
       break
cap.release()
cv2.destroyAllWindows()
```

# OBJECT DETECTION PLUS LIVE VIDEO PLUS SAVE VIDEO PLUS SHOW TOTAL OBJECTS DETECTED 

```python
import cv2
from ultralytics import YOLO

model=YOLO('yolov8n.pt')
cap=cv2.VideoCapture(0)
frame_width=cap.get(cv2.CAP_PROP_FRAME_WIDTH)
frame_height=cap.get(cv2.CAP_PROP_FRAME_HEIGHT)
fps=cap.get(cv2.CAP_PROP_FPS)

fourcc=cv2.VideoWriter_fourcc(*'mp4v')

out=cv2.VideoWriter(r'E:\video2.mp4',fourcc,int(fps),(int(frame_width),int(frame_height)))

while cap.isOpened():
    OPEN,frame=cap.read()
    if not OPEN:
        break
    results=model(frame)
    for i in results:
        bbox=i.boxes.xyxy.numpy()
        confs=i.boxes.conf.numpy()
        class_ids=i.boxes.cls.numpy()
        
    total_objects=len(bbox)
    
    for a,b,c in zip(bbox,confs,class_ids):
        x1,y1,x2,y2=map(int,a)
        
        cv2.rectangle(frame,(x1,y1),(x2,y2), (0,255,0),1)
        cv2.putText(frame, model.names[c],(x1,y1-20),cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0,0,0),2)
        label=f'TOTAL OBJECTS:{total_objects}'
        cv2.putText(frame,label, (10,30), cv2.FONT_HERSHEY_SIMPLEX,0.5,(0,255,0),2)
        
    out.write(frame)
    
    cv2.imshow('video2',frame)
    if cv2.waitKey(1) & 0xFF==ord('q'):
        break
cap.release()
out.release()
cv2.destroyAllWindows()
```

```python
input_video=r'E:\new_downloads_default_folder\12.mp4'
cap=cv2.VideoCapture(0)

frame_width=cap.get(cv2.CAP_PROP_FRAME_WIDTH)
frame_height=cap.get(cv2.CAP_PROP_FRAME_HEIGHT)
fps=cap.get(cv2.CAP_PROP_FPS)

fourcc=cv2.VideoWriter_fourcc(*'mp4v')

out=cv2.VideoWriter(r'E:\new_downloads_default_folder\counting.mp4',fourcc,int(fps),(int(frame_width),int(frame_height)))

while cap.isOpened():
    OPEN,frame=cap.read()
    if not OPEN:
        break
    results=model(frame)
    
    for i in results:
        bbox=i.boxes.xyxy.numpy()
        confs=i.boxes.conf.numpy()
        class_ids=i.boxes.cls.numpy()
    count=1
    total_objects=len(bbox)

    for a,b,c in zip(bbox,confs,class_ids):
        x1,y1,x2,y2=map(int,a)
        if b<0.33:
            break
        label1=f'{model.names[c]}:{count}[ {b*100:.0f}%]'         
        
        cv2.rectangle(frame,(x1,y1),(x2,y2), (0,255,0),2)
        cv2.putText(frame, label1,(x1,y1-20),cv2.FONT_HERSHEY_SIMPLEX, 0.75, (0,0,0),2)
        
        label2=f'TOTAL OBJECTS:{total_objects}'
        
        cv2.putText(frame,label2, (41,41), cv2.FONT_HERSHEY_SIMPLEX,1.5,(0,255,0),5)
        count=count+1
    
    out.write(frame)
    
    cv2.imshow('video2',frame)
    if cv2.waitKey(1) & 0xFF==ord('q'):
        break
cap.release()
out.release()
cv2.destroyAllWindows()
```
